```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IMDB Dataset.csv")
```

```python
data
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
reviews = data['review']
```

```python
voc_size = 10000
```

### One hot Encoding

```python
onehot_rep = [one_hot(sentence, voc_size) for sentence in reviews]
print(onehot_rep)
```

### Word Embedding Representaion

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
## pre-padding
sent_length = 400
embedded_docs = pad_sequences(onehot_rep, padding='pre', maxlen = sent_length)
print(embedded_docs)
```

```python
embedded_docs[21212]
```

```python
dim = 50
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim))
model.build(input_shape=(None, sent_length))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(np.expand_dims(embedded_docs[0], axis=0))
```

```python
print(model.predict(embedded_docs))
```
